```python
# Tickers: High Volume Liquid Stocks and ETFs preferably. Emphasis on low lost options near the money.
# This list can be modified to include more tickers or different ones based on your trading strategy.
tickers = [
    'AAPL', 'MSFT', 'NVDA', 'AMZN', 'META', 'TSLA', 'AMD', 'GOOGL', 'NFLX', 'SHOP', 'COIN', 'TSM',
    'INTC', 'QCOM', 'BA', 'JPM', 'XOM', 'WMT', 'COST', 'DIS', 'NKE', 'MCD', 'ORCL', 'SPY', 'QQQ', 'TLT',
    'F', 'GM', 'PLTR', 'DKNG', 'UBER', 'ABNB', 'MARA', 'ARKK', 'MU', 'SNOW', 'ADBE', 'GME', 'AMC', 'PYPL',
    'PFE', 'MRK', 'UNH', 'ABBV', 'TMO', 'LLY','TFC','SCHD', 'AVGO', 'UBER','CSCO','SPLG','XLF','XLE','XLV',
    'BAC','CCL','IWM','AMD','SPXU','SPXS','SPXL','SPYV','SPYG','QQQM','QQQJ','VTI','VOO',
    'VUG', 'VTV', 'IWD', 'IWF', 'IWM', 'IWB', 'IVV', 'VO', 'VBK', 'VBR', 'VIOO', 'VTWO', 'VTWO',
]
```

```python
#Example to get options chain for a specific stock and expiration date
symbol = 'AAPL'
expiration = '2025-06-06'
chain2 = tradier.market.get_option_chains(symbol, expiration, greeks=True)
chain2
```

```python
results = []

for symbol in tickers:
        # Get all expirations for the symbol
        expirations_df = tradier.market.get_option_expirations(symbol)
        expirations = expirations_df.index.tolist()
        
        valid_exp = next(
            (exp for exp in expirations
             if datetime.strptime(str(exp), "%Y-%m-%d") > datetime.now() + timedelta(days=30)),
            None
        )

        if not valid_exp:
            print(f"⚠️ No valid expiration for {symbol}")
            continue

        # Get option chain for the valid expiration
        chain = tradier.market.get_option_chains(symbol, valid_exp, greeks=True)

        for _, opt in chain.iterrows():
                if (
                    # Set criteria for filtering options
                    opt['option_type'] == 'call' and
                    opt['greeks.delta'] is not None and opt['greeks.delta'] >= 0.10 and
                    opt['ask'] is not None and opt['ask'] <= 0.25 and
                    opt['volume'] is not None and opt['volume'] >= 100 and
                    opt['greeks.smv_vol'] is not None and opt['greeks.smv_vol'] <= 0.6
                ):
                    results.append({
                        'Ticker': symbol,
                        'Symbol': opt['symbol'],
                        'Strike': opt['strike'],
                        'Expiration': opt['expiration_date'],
                        'Delta': round(opt['greeks.delta'], 2),
                        'IV': round(opt['greeks.smv_vol'], 2),
                        'Ask': opt['ask'],
                        'Volume': opt['volume'],
                        'Underlying Price': opt.get('underlying_price'),
                        'Open Interest': opt['open_interest']
                    })

# Save to DataFrame
df = pd.DataFrame(results)
if not df.empty:
    df.sort_values(by='Delta', ascending=False, inplace=True)
    df.to_csv('filtered_options.csv', index=False)
    print(f"\n✅ Screener complete. {len(df)} options saved to 'filtered_options.csv'")
else:
    print("\n⚠️ No options met your criteria.")
```
